```python
# prompt: buatkan perintah crud

from pymongo import MongoClient

# Ganti dengan connection string yang Anda dapatkan
uri = "mongodb+srv://noufalaji4:noufalaji4@cluster0.y6qah.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

# Inisialisasi klien dengan URI
client = MongoClient(uri)

# Mengakses database yang dituju
db = client['company_db']  # Ganti dengan nama database Anda

# Mengakses koleksi (collection) dalam database
collection = db['employees']  # Ganti dengan nama koleksi Anda


def create_employee(employee_data):
  """
  Membuat data karyawan baru.

  Args:
    employee_data: Dictionary berisi data karyawan.
  """
  result = collection.insert_one(employee_data)
  print(f"Employee created with ID: {result.inserted_id}")


def read_employees(query={}):
  """
  Membaca data karyawan berdasarkan query.

  Args:
    query: Dictionary berisi kriteria pencarian. Jika kosong, semua karyawan akan ditampilkan.
  """
  for employee in collection.find(query):
    print(employee)


def update_employee(query, new_data):
  """
  Memperbarui data karyawan berdasarkan query.

  Args:
    query: Dictionary berisi kriteria pencarian.
    new_data: Dictionary berisi data yang akan diperbarui.
  """
  result = collection.update_one(query, {"$set": new_data})
  print(f"{result.modified_count} employee(s) updated.")


def delete_employee(query):
  """
  Menghapus data karyawan berdasarkan query.

  Args:
    query: Dictionary berisi kriteria pencarian.
  """
  result = collection.delete_one(query)
  print(f"{result.deleted_count} employee(s) deleted.")


# Contoh Penggunaan:

# Create a new employee
new_employee = {
    "name": "John Doe",
    "department": "Sales",
    "salary": 60000
}
create_employee(new_employee)

# Read all employees
print("\nAll employees:")
read_employees()

# Read employee with specific name
print("\nEmployee with name 'John Doe':")
read_employees({"name": "John Doe"})


# Update employee's salary
update_employee({"name": "John Doe"}, {"salary": 65000})

# Delete employee
delete_employee({"name": "John Doe"})

# Read all employees again
print("\nAll employees after deletion:")
read_employees()
```

```python
# Buat koleksi baru 'products'
products_collection = db['products']

products = [
    {'name': 'Kunci Pas', 'brand': 'Stanley', 'price': 150, 'stock': 80},
    {'name': 'Radiator', 'brand': 'Denso', 'price': 1200, 'stock': 20},
    {'name': 'Pompa Oli', 'brand': 'Lisle', 'price': 300, 'stock': 50},
    {'name': 'Alat Pengukur Tekanan Ban', 'brand': 'Slime', 'price': 100, 'stock': 100},
    {'name': 'Kompresor Udara', 'brand': ' PORTER-CABLE', 'price': 2500, 'stock': 15},
    {'name': 'Pembersih Karburator', 'brand': 'Gunk', 'price': 200, 'stock': 60},
    {'name': 'Penyedot Debu Mobil', 'brand': 'Black+Decker', 'price': 800, 'stock': 30},
    {'name': 'Kit Perbaikan Ban', 'brand': 'Slime', 'price': 150, 'stock': 50},
    {'name': 'Pengangkat Mobil ', 'brand': 'Denso', 'price': 600, 'stock': 25},
    {'name': 'Kain Pembersih Mobil', 'brand': 'Denso', 'price': 75, 'stock': 90}
]

# Menyisipkan data produk bengkel mobil ke dalam koleksi
result = products_collection.insert_many(products)
print(f'Data produk bengkel mobil telah dimasukkan dengan ID: {result.inserted_ids}')
```

```python
# Menampilkan semua produk dalam koleksi
print('Daftar semua produk yang telah dimasukkan:')
all_products = products_collection.find()  # Mengambil semua dokumen dari koleksi

# Menampilkan data produk
for product in all_products:
    print(product)
```

```python
average_price_result = products_collection.aggregate([
    { '$group': {  '_id': None, 'averagePrice': {'$avg': '$price'}}}  ])

# Mengambil nilai rata-rata
average_price = next(average_price_result)['averagePrice']

# Mencari produk dengan harga di atas rata-rata
expensive_products = list(products_collection.find({'price': {'$gt': average_price}}))

# Menampilkan hasil produk yang harganya di atas rata-rata
print(f'Harga rata-rata: {average_price}')
print('Produk dengan harga di atas rata-rata:')
for product in expensive_products:
    print(product)
```

```python
# Aggregation pipeline untuk menghitung total produk dalam setiap kategori
pipeline = [{  '$group': {'_id': '$category',  # Kelompokkan berdasarkan kategori
 'totalProducts': {'$sum': '$stock'}  # Hitung total stok untuk setiap kategori
}},  {'$sort': {'totalProducts': -1  # Urutkan berdasarkan total produk (descending)
}} ]
# Menjalankan agregasi
category_totals = list(products_collection.aggregate(pipeline))

# Menampilkan hasil
print('Total produk dalam setiap kategori:')
for category in category_totals:
    print(f'Kategori: {category["_id"]}, Total Produk: {category["totalProducts"]}')
```

```python
# Aggregation pipeline untuk menemukan 5 produk dengan harga tertinggi per brand
pipeline = [
    {
        '$sort': {
            'price': -1  # Urutkan berdasarkan harga tertinggi
        }
    },
    {
        '$group': {
            '_id': '$brand',  # Kelompokkan berdasarkan brand
            'top_products': {
                '$push': {
                    'name': '$name',
                    'price': '$price',
                    'stock': '$stock'
                }
            }
        }
    },
    {
        '$project': {
            'brand': '$_id',
            'top_products': {'$slice': ['$top_products', 5]}  # Ambil 5 produk teratas
        }
    }
]

# Menjalankan agregasi
top_products_per_brand = list(products_collection.aggregate(pipeline))

# Menampilkan hasil
print('5 Produk dengan Harga Tertinggi per Brand:')
for brand in top_products_per_brand:
    print(f'Brand: {brand["brand"]}')
    for product in brand['top_products']:
        print(f'  - Produk: {product["name"]}, Harga: {product["price"]}, Stok: {product["stock"]}')
```

```python
# Menghitung total price dan rata-rata stok per brand
pipeline = [
    {
        '$group': {
            '_id': '$brand',
            'total_price': {'$sum': '$price'},
            'average_stock': {'$avg': '$stock'}
        }
    },
    {
        '$sort': {'_id': 1}  # Mengurutkan berdasarkan brand
    }
]

# Menjalankan agregasi
report = products_collection.aggregate(pipeline)

# Menampilkan laporan
print("Laporan Total Harga dan Rata-rata Stok per Brand:")
for item in report:
    print(f"Brand: {item['_id']}, Total Price: {item['total_price']}, Average Stock: {item['average_stock']}")
```
