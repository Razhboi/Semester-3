```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

```python
# Exercise 2: Standardizing Categorical Data
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import matplotlib.pyplot as plt

# Data for plotting
categories = ['Category A', 'Category B', 'Category C']
values = [23, 17, 35]

# Plotting a pie chart
plt.figure(figsize=(8, 5))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=['cyan', 'magenta', 'yellow'])
plt.title('Pie Chart of Categories')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd

# Muat dataset
df = pd.read_csv('tsunami_dataset.csv')
```

```python
# Menghitung jumlah nilai yang hilang di setiap kolom
print(df.isnull().sum())
```

```python
# Mengisi missing values
df['MINUTE'].fillna(df['MINUTE'].mean(), inplace=True)  # Mengisi missing values di MINUTE dengan rata-rata
df['MONTH'].fillna(df['MONTH'].median(), inplace=True)    # Mengisi missing values di MONTH dengan median

# Menghapus baris dengan missing values di kolom CAUSE
df.dropna(subset=['CAUSE'], inplace=True)

# Menampilkan hasil setelah pembersihan
print('Setelah pembersihan:\n', df)
```

```python
# Standardize category values
df['LOCATION_NAME'] = df['LOCATION_NAME'].str.lower()
print('Standardized Data:\n', df)
```

```python
# Convert 'MONTH' and 'DAY' columns to appropriate data types
df['MONTH'] = df['MONTH'].astype(int)
df['DAY'] = df['DAY'].astype(int)

# Menghitung jumlah tsunami per bulan
monthly_counts = df.groupby('MONTH')['DAY'].count()

# Plotting
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='cyan')  # Diagram batang
plt.title('Jumlah Tsunami per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Tsunami')
plt.xticks(rotation=0)  # Menjaga label pada sumbu X tetap horizontal
plt.grid(axis='y')  # Menambahkan grid pada sumbu Y
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('tsunami_dataset.csv')

# Drop rows with NaN values in 'MONTH' column
df = df.dropna(subset=['MONTH'])

# Convert 'MONTH' to appropriate data type
df['MONTH'] = df['MONTH'].astype(int)

# Hitung jumlah tsunami per bulan
monthly_counts = df['MONTH'].value_counts().sort_index()
months = monthly_counts.index.values.reshape(-1, 1)
counts = monthly_counts.values

# Linear regression
model = LinearRegression()
model.fit(months, counts)
predictions = model.predict(months)

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(months, counts, color='dodgerblue', label='Data Tsunami')  # Scatter plot
plt.plot(months, predictions, color='blue', label='Regresi Linier')  # Regresi linier
plt.fill_between(months.flatten(), predictions - 5, predictions + 5, color='lightblue', alpha=0.4, label='Interval Kepercayaan')  # Interval kepercayaan

# Menambahkan label dan judul
plt.title('Hubungan Antara Bulan dan Jumlah Tsunami')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Tsunami')
plt.xticks(months.flatten(), rotation=0)  # Label bulan horizontal
plt.legend()
plt.grid(True)
plt.show()
```
